### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular el ingreso total de cada tienda
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()
```

```python
# Imprimir los resultados para cada tienda
print(f"Ingreso total de la tienda 1: {ingreso_total_tienda1}")
print(f"Ingreso total de la tienda 2: {ingreso_total_tienda2}")
print(f"Ingreso total de la tienda 3: {ingreso_total_tienda3}")
print(f"Ingreso total de la tienda 4: {ingreso_total_tienda4}")
```

# 2. Ventas por categoría

```python
# Contar productos por categoría en la tienda 1
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
print("Ventas por categoría en la Tienda 1:")
print(ventas_por_categoria_tienda1)
print("\n" + "="*30 + "\n")

# Contar productos por categoría en la tienda 2
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
print("Ventas por categoría en la Tienda 2:")
print(ventas_por_categoria_tienda2)
print("\n" + "="*30 + "\n")

# Contar productos por categoría en la tienda 3
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
print("Ventas por categoría en la Tienda 3:")
print(ventas_por_categoria_tienda3)
print("\n" + "="*30 + "\n")

# Contar productos por categoría en la tienda 4
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()
print("Ventas por categoría en la Tienda 4:")
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio de cada tienda
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

# Imprimir los resultados
print(f"Calificación promedio de la Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio de la Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio de la Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio de la Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Crear una lista con todos los DataFrames de las tiendas
tiendas_dfs = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
colores = ['skyblue', 'lightgreen', 'salmon', 'gold']

# Iterar sobre la lista de DataFrames y nombres
for i, df in enumerate(tiendas_dfs):
    # Contar las ventas de cada producto
    ventas_productos = df['Producto'].value_counts()

    # Obtener los 5 más y menos vendidos
    mas_vendidos = ventas_productos.head(5)
    menos_vendidos = ventas_productos.tail(5)

    print(f"--- {nombres_tiendas[i]} ---")
    print(f"Productos más vendidos:")
    print(mas_vendidos)
    print("\nProductos menos vendidos:")
    print(menos_vendidos)

    # Crear un gráfico de barras para los productos más vendidos
    plt.figure(figsize=(10, 6))
    mas_vendidos.plot(kind='bar', color=colores[i])
    plt.title(f'Top 5 Productos Más Vendidos - {nombres_tiendas[i]}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad de Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    print("\n" + "="*50 + "\n")
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio para cada tienda
envio_promedio_tienda1 = tienda['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

# Imprimir los resultados
print(f"Costo de envío promedio de la Tienda 1: {envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio de la Tienda 2: {envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio de la Tienda 3: {envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio de la Tienda 4: {envio_promedio_tienda4:.2f}")
```
